To speed-up setup and avoid memory limits we only display 1-in-4 images

```python
%pylab notebook
from pycheops import Dataset
from tqdm import tqdm_notebook as tqdm
```

```python
ds=Dataset.from_test_data('DP-079', target='HD97658')
cube = ds.get_imagettes()
cube1in4 = (cube[::4,:,:])
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
fig = plt.figure()
ims = []
for i in tqdm(range(len(cube1in4))):
    im = plt.imshow(cube1in4[i,:,:],interpolation='nearest',cmap='gist_heat', animated=True)
    ims.append([im])
anim = animation.ArtistAnimation(fig, ims, interval=100, blit=True)
```
