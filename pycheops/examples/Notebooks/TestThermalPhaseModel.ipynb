```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pycheops.models import FactorModel, EclipseModel, ThermalPhaseModel
from lmfit import report_fit
```

```python
T_0 = 0.5
P = 2.0
a_th = 0.012
b_th = -0.003
c_th = 0.0027
D = 0.04
W = 0.06
b = 0.1
L = 0.001
```

```python
m = ThermalPhaseModel(prefix="T")*EclipseModel()*FactorModel()
pars = m.make_params(T_0=T_0, 
                     Ta_th=a_th, Tb_th=b_th, Tc_th=c_th,
                     D=D, W=W, b=b, L=L)
pars['P'].set(value=P, vary=False)
pars['TT_0'].set(expr='T_0')
pars['TP'].set(expr='P')
pars
```

```python
t = np.arange(T_0,T_0 + P, 0.001)
yerr = np.full_like(t, 450e-6)
flux_true = m.eval(params=pars, t=t)
flux = flux_true + np.random.normal(0.0, yerr, size=len(t))
fig,ax=plt.subplots()
ax.scatter(t, flux,s=4)
ax.plot(t, flux_true,'r')
ax.set_xlabel('Time')
ax.set_ylabel('Flux')
ax.set_xlim(min(t),max(t));
```

```python
print('Max flux from model = {:0.4f}, F = {:0.4f}'.format(np.max(flux_true),pars['TFmax'].value))
print('Phase of maximum flux from model = {:0.4f}, ph_max = {:0.4f}'
      .format((t[np.argmax(flux_true)]-T_0)/P,pars['Tph_max'].value % 1))
print('Peak-to-trough amplitude from model = {:0.3f}, A = {:0.3f}'
      .format(np.max(flux_true)-np.min(flux_true),pars['TA'].value))
```

```python
result = m.fit(flux, pars, t=t, weights=1/yerr)
report_fit(result)
```

```python
result.plot(xlabel='Time',ylabel='Flux',
            data_kws={'ms':1,'alpha':0.2},
            fit_kws={'c':'r'},
            ax_fit_kws={'xlim':[min(t),max(t)]})
```
